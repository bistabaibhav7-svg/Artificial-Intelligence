```python
# Problem 1 - Sorting:
# 3.1 Warm Up Exercises:
# 1. Sorting and Subsetting:
# Complete all following Task:
# â€¢ Dataset for the Task: "titanic.csv"
# Following task is common for all the problem:
# 1. Load the provided dataset and import in pandas DataFrame.
# 2. Check info of the DataFrame and identify following:

import pandas as pd

 # Load the dataset
titanic_df = pd.read_csv('Titanic-Dataset.csv')

# Check info of the DataFrame
print("Dataset Info:")
print(titanic_df.info())
print("\nFirst 5 rows:")
print(titanic_df.head())

#1. Create a DataFrame called fare that contains only the Fare column
fare = titanic_df[['Fare']]
print("Fare DataFrame (first 5 rows):")
print(fare.head())

#2. Create a DataFrame called class_age that contains Pclass and Age columns
class_age = titanic_df[['Pclass', 'Age']]
print("\nClass and Age DataFrame (first 5 rows):")
print(class_age.head())

#3. Create a DataFrame called survived_gender that contains Survived and Sex columns
survived_gender = titanic_df[['Survived', 'Sex']]
print("\nSurvived and Gender DataFrame (first 5 rows):")
print(survived_gender.head())
```

```python
# Problem 2 - Subsetting
# Subsetting Rows:
# 1. Filter for fare > 100
fare_gt_100 = titanic_df[titanic_df['Fare'] > 100]
print(f"Passengers with fare > 100: {len(fare_gt_100)}")
print("First 5 rows:")
print(fare_gt_100.head())

# 2. Filter for first class passengers (Pclass = 1)
first_class = titanic_df[titanic_df['Pclass'] == 1]
print(f"\nFirst class passengers: {len(first_class)}")
print("First 5 rows:")
print(first_class.head())

# 3. Filter for female passengers under 18
female_under_18 = titanic_df[(titanic_df['Age'] < 18) & (titanic_df['Sex'] == 'female')]
print(f"\nFemale passengers under 18: {len(female_under_18)}")
print("First 5 rows:")
print(female_under_18.head())
```

```python
#Subsetting Rows by Categorical variables:
# 1. Filter for passengers embarked at 'C' or 'S'
embarked_c_or_s = titanic_df[titanic_df['Embarked'].isin(['C', 'S'])]
print(f"Passengers embarked at C or S: {len(embarked_c_or_s)}")
print("First 5 rows:")
print(embarked_c_or_s.head())

# 2. Filter for first or second class passengers
first_second_class = titanic_df[titanic_df['Pclass'].isin([1, 2])]
print(f"\nFirst or second class passengers: {len(first_second_class)}")
print("First 5 rows:")
print(first_second_class.head())
```

```python
# 3.2 Exploratory Data Analysis Practice Exercise - 1
# Which passenger had the highest fare paid relative to their age?
# Create a copy of the dataframe to avoid SettingWithCopyWarning
titanic_eda = titanic_df.copy()

# Handle missing values in Age column by filling with median age
median_age = titanic_eda['Age'].median()
titanic_eda['Age'] = titanic_eda['Age'].fillna(median_age)

# Add fare_per_year column
titanic_eda['fare_per_year'] = titanic_eda['Fare'] / titanic_eda['Age']

# Subset rows where fare_per_year > 5
high_fare_age = titanic_eda[titanic_eda['fare_per_year'] > 5]

# Sort by descending fare_per_year
high_fare_age_srt = high_fare_age.sort_values('fare_per_year', ascending=False)

# Select only Name and fare_per_year columns
result = high_fare_age_srt[['Name', 'fare_per_year']]

print("Passengers with highest fare per year (first 10):")
print(result.head(10))

print(f"\nThe passenger with highest fare relative to age is: {result.iloc[0]['Name']}")
print(f"With fare per year of: {result.iloc[0]['fare_per_year']:.2f}")
```

```python
# Which adult male passenger paid the highest fare relative to their class?
# Add fare_per_class column
titanic_eda['fare_per_class'] = titanic_eda['Fare'] / titanic_eda['Pclass']

# Subset for adult male passengers
adult_males = titanic_eda[(titanic_eda['Sex'] == 'male') & (titanic_eda['Age'] >= 18)]

# Sort by descending fare_per_class
adult_males_srt = adult_males.sort_values('fare_per_class', ascending=False)

# Select required columns
result_males = adult_males_srt[['Name', 'Age', 'fare_per_class']]

print("Adult male passengers with highest fare per class (first 10):")
print(result_males.head(10))

print(f"\nThe adult male passenger with highest fare relative to class is: {result_males.iloc[0]['Name']}")
print(f"Age: {result_males.iloc[0]['Age']}, Fare per class: {result_males.iloc[0]['fare_per_class']:.2f}")
```

```python
# 3.3 Exploratory Data Analysis with Group-by Method Practice Exercise
# What percent of the total fare revenue came from each passenger class?
# Calculate total fare
total_fare = titanic_df['Fare'].sum()
print(f"Total fare revenue: {total_fare:.2f}")

# Calculate fare by class
fare_by_class = titanic_df.groupby('Pclass')['Fare'].sum()
print("\nFare revenue by class:")
print(fare_by_class)

# Calculate percentage
percentage_by_class = (fare_by_class / total_fare) * 100
print("\nPercentage of total fare revenue by class:")
for pclass, percentage in percentage_by_class.items():
    print(f"Class {pclass}: {percentage:.2f}%")
```

```python
# What percent of the total number of passengers belonged to each age group?
# Create age_group column
def categorize_age(age):
    if pd.isna(age):
        return 'Unknown'
    elif age < 18:
        return 'Child'
    elif age < 65:
        return 'Adult'
    else:
        return 'Senior'

titanic_df['age_group'] = titanic_df['Age'].apply(categorize_age)

# Calculate total number of passengers
total_passengers = len(titanic_df)
print(f"Total passengers: {total_passengers}")

# Count passengers in each age group
age_group_counts = titanic_df['age_group'].value_counts()
print("\nPassengers by age group:")
print(age_group_counts)

# Calculate percentage
age_group_percentage = (age_group_counts / total_passengers) * 100
print("\nPercentage of passengers by age group:")
for group, percentage in age_group_percentage.items():
    print(f"{group}: {percentage:.2f}%")

```
